#### Import Liabraries

```python
import pandas as pd
import numpy as np
import os
```

```python
pwd= os.getcwd()
```

```python
cyclistic_data= pd.read_csv(pwd+ "\\Cyclistic_bike_ride_data_2022.csv")
cyclistic_data.head()
```

#### Taking sample

```python
list= list(cyclistic_data)   # converting to a list
sample_size= int(len(cyclistic_data) * 0.1)

cyclistic_sample= cyclistic_data.sample(n= sample_size, random_state= 1)
print(cyclistic_sample.shape)
```

```python
print("real data:")
print(cyclistic_data.describe())
print("")
print("sample:")
print(cyclistic_sample.describe())
```

```python
cyclistic_sample.head()
```

#### Data Cleaning

```python
cyclistic_sample.dtypes.to_frame()
```

```python
# Count NaN values

cyclistic_sample_NaN_counts= cyclistic_sample.isna().sum(axis=0)
cyclistic_sample_NaN_counts
```

```python
# Recounting NaN values

cyclistic_sample_NaN_counts= cyclistic_sample.isna().sum(axis=0)
cyclistic_sample_NaN_counts
```

```python
## drop NaN values

cyclistic_sample= cyclistic_sample.dropna()
print(cyclistic_sample.shape)
```

```python
# Changing data types

cyclistic_sample[["started_at", "ended_at"]]= cyclistic_sample[["started_at", "ended_at"]].astype("datetime64")
cyclistic_sample.dtypes.to_frame()
```

here

```python
## creating new column from started_at and ended_at columns

cyclistic_sample["ride_duration"]= cyclistic_sample["ended_at"] - cyclistic_sample["started_at"]
cyclistic_sample.head()
```

```python
## creating another column containing only time of ride duration

cyclistic_sample["ride_duration"]= cyclistic_sample["ride_duration"].astype(str)
cyclistic_sample["ride_length"] = cyclistic_sample['ride_duration'].str[-8:]
cyclistic_sample.head()
```

```python
# creating weekday column contains the day of the week 

cyclistic_sample["weekday"]= cyclistic_sample["started_at"].dt.weekday + 1
cyclistic_sample.head()
```

```python
# create new column named ride_length_minutes

cyclistic_sample["ride_length_minutes"]= (pd.to_datetime(cyclistic_sample["ended_at"]) - pd.to_datetime(cyclistic_sample["started_at"])).dt.total_seconds() / 60

# create new column named ride_length_hours

cyclistic_sample["ride_length_hours"]= cyclistic_sample["ride_length_minutes"] / 60

# make the hours column containing two decimal points

cyclistic_sample['ride_length_hours']= round(cyclistic_sample['ride_length_hours'], 2)

cyclistic_sample.head(3)
```

```python
# Create a new column called member_type

cyclistic_sample["member_type"]= cyclistic_sample["member_casual"].apply(lambda x: "Annual Member" if x== "member" else "Casual Rider")

cyclistic_sample.head()
```

```python
## Changing datatype of started_at column

cyclistic_sample["started_at"]= pd.to_datetime(cyclistic_sample["started_at"])
```

```python
cyclistic_sample["month_name"]= cyclistic_sample["started_at"].apply(lambda x: x.strftime('%B'))
cyclistic_sample.head(3)
```

```python
## Get the hour time in 24 hour format

cyclistic_sample["hour_time"]= cyclistic_sample["started_at"].apply(lambda x: x.strftime('%H%M') + "hours")
cyclistic_sample.head(3)
```

```python
# save cleaned data

cyclistic_sample.to_csv("cleaned_cyclistic_data.csv", index= False)
```
